```python
import pandas as pd
```

```python
df = pd.read_csv('loan.csv')
```

```python
df
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoders = {}
for column in ['gender', 'occupation', 'education_level', 'marital_status', 'loan_status']:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('loan_status', axis=1)
y = df['loan_status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python

y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
print(f'Accuracy: {accuracy}')
print('Classification Report:')
print(report)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(30, 20))
plot_tree(model, feature_names=X.columns, class_names=label_encoders['loan_status'].classes_, fontsize=12, filled=True)
plt.title("Decision Tree Visualization", fontsize=24)
plt.savefig('decision_tree_visualization.png')  # Save the plot as a PNG file
plt.close() 
```
